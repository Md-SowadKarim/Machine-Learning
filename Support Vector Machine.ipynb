```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('credit card.csv')
df
```

# Shape view

```python
df.shape
```

#  NUll Value Checking

```python
df.isnull().sum()
```

# Null Value Handaling

```python
handle=df['AGE'].mean()
handle
```

```python
df.AGE=df.AGE.fillna(handle)
df
```

# Initializing value of x & y

```python
x=df.drop(['default.payment.next.month'],axis=1)
x
```

```python
y=df['default.payment.next.month']
y
```

# Spliting Data For Train & Test

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train
```

```python
y_test
```

# Model With Support vectior machine (SVC)

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Model with Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=50) # with n_estimatiors=(500)tree number
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

# Model with Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_test,y_test)
```

# Naive Bayes

# Bernulli ======

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
bnb.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
mnb.score(x_test,y_test)
```
